# Numpy - Array
*  NumPy’s main object is the homogeneous multidimensional array. It is a table of elements (usually numbers), all of the same type, indexed by a tuple of non-negative integers. In NumPy dimensions are called axes.
* NumPy’s array class is called ndarray. It is also known by the alias array. Note that numpy.array is not the same as the Standard Python Library class array.array, which only handles one-dimensional arrays and offers less functionality.

#### Intrinsic NumPy Array Creation

###### np.zeros
>* zeros(shape) will create an array filled with 0 values with the specified shape. The default dtype is float64.

```python
zeroes=np.zeros((2, 3))
zeroes
```

###### np.arrange()
>* arange() will create arrays with regularly incrementing values

```python
np.arange(10)
```

```python
np.arange(2, 10, dtype=float)
```

```python
np.arange(2, 3, 0.1)
```

##### np.linspace()
linspace() will create arrays with a specified number of elements, and spaced equally between the specified beginning and end values

```python
x=np.linspace(2.0, 3.0, num=12)
x
```

```python
x.reshape(6,2)
```

```python
x.reshape(4,3)
```

```python
x.reshape(5,3)
```

##### Operations on Numpy Arrays

If an operation called once is applied to all the elements, it is called broadcasting. 

```python
a = np.array([[1,2],[3,4]])
b = np.array([[5,6],[7,8]])
```

```python
a
```

```python
a+1
```

```python
a+b
```

```python
a.sum(axis=1)
```

```python
a.sum() # Numpy sum() function for numpy arrays
```

```python
sum(a) ## Default Python sum function on integers
```

```python
a
```

```python
a.max() # Get the max Value
```

```python
a.min() # # Get the minimum value
```

```python
a.argmin() # Get index of the minimum value
```

##### Image as Numpy Array

An image is a 3D Array... Why 3D?

Imaging a 2D array where each pixel is a number between 0 to 255. 0 is black, 255 is white, other numbers are shades of gray. \


![image.png](image.png)

In full color image, we need three channels - viz, Red, Green and Blue. So instead of having one value at each pixel, we have three values. 

So a 600x400 pixel image means:
- 400 rows
- 600 columns, thus cell indicating a pixel
- 3 channels at each cell
