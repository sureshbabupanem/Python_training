```python
class JobSeeker():
    """
    JobSeeker is an abstraction for a current or a completed studend who is looking for a job. The student might already be
    working. 
    
    Attributes:
    name: Name of the Candidate
    certificates: List containing attained certificates in string
    skills: List containing skills for the topic classes have been attended
    programming_experience: int containing total number of hours of programming practice, assignment or development
    """
    def __init__(self, name, programming_experience):
        self.name = name
        self.programming_experience = programming_experience
        self.certificates = []
        self.skills = []
        
    def attend_class ( self, instructor, topic, hours ):
        """
        Add skill based on the topic to the jobseeker's profile
        
        Parameters:
        instructor: Name of the trainer (str)
        topic: Name of the topic as str, will be added to the skill list
        hours: Number of hours, int
        """
        self.skills.append(topic)
        
    def attain_certificate( self, certificate ):
        self.certificates.append (certificate)
        
    def get_name(self):
        return self.name
        
    def give_description( self ):
        print ("Name: "+ self.name)
        print ("Certificate: ")
        if len(self.certificates) == 0:
            print ('No Certificates Yet')
        else:
            print ("Certificate:",self.certificates)
        print ("Skills: ",(self.skills) )
        print ("Programming Experience: "+str(self.programming_experience)+ " hours")
    
```

```python
mohit = JobSeeker("Mohit", 100)
akshay = JobSeeker("Akshay", 150)
```

```python
akshay.get_name()
```

```python
mohit.get_name()
```

```python
mohit.attend_class('Aditya', 'Classes and Objects', 2)
mohit.attain_certificate('IBM Certified Data Scientist')
mohit.give_description()
```

## Zipping Files in Python
ZIP is an archive file format that supports lossless data compression. By lossless compression, we mean that the compression algorithm allows the original data to be perfectly reconstructed from the compressed data. So, a ZIP file is a single file containing one or more compressed files, offering an ideal way to make large files smaller and keep related files together.

We will use an inbuilt python module called zipfile.

Let's check the documentation before zipping:
https://docs.python.org/2/library/zipfile.html

**Note**: The aim of this section is to introduce you to the idea of using external modules in Python. You do not have to remember what we will see here - so focus on just understanding how we use a library.

```python
import zipfile
```

```python
files_to_zip = [ 'test.txt', 'test1.txt', 'temp.txt' ]


zf = zipfile.ZipFile('zipfile_write.zip', mode='w')


for filename in files_to_zip:
    try:
        zf.write(filename)
    except FileNotFoundError:
        print (filename+" had some issue")


zf.close()
```
